```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

# 
file_path = '/home/zhangyuhao/Desktop/Result/Online_Move_Correct/Spectrum_analysis/'
data = np.load(file_path + '/20230510/Interposed nucleus_stan_out_eigvc.npy')  #20230511 20230513 20230516 20230517 20230518 20230519 20230521 20230522

# 
cos_sim_matrix = cosine_similarity(data)

# 
euclidean_matrix = euclidean_distances(data)

# PCA  2D
pca_2d = PCA(n_components=2)
data_pca_2d = pca_2d.fit_transform(data)

# 
plt.figure(figsize=(6, 6))
plt.imshow(cos_sim_matrix, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Cosine Similarity')
plt.title('Cosine Similarity Heatmap')
plt.xlabel('Vector Index')
plt.ylabel('Vector Index')
plt.tight_layout()
plt.show()

# 
plt.figure(figsize=(6, 6))
plt.imshow(euclidean_matrix, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Euclidean Distance')
plt.title('Euclidean Distance Heatmap')
plt.xlabel('Vector Index')
plt.ylabel('Vector Index')
plt.tight_layout()
plt.show()

#  PCA 2D 
plt.figure(figsize=(6, 6))
plt.scatter(data_pca_2d[:, 0], data_pca_2d[:, 1], c='blue', alpha=0.7, label='Vectors')
plt.title('PCA Projection (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.tight_layout()
plt.show()

#  PCA 
explained_variance_ratio = pca_2d.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance_ratio}")
```
